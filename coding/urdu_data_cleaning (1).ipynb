```python
import re
import emoji

# Define a function to remove emojis from text
def remove_emojis(text):
    # Remove emojis using the emoji library
    clean_text = emoji.demojize(text)
    # Remove URLs
    clean_text = re.sub(r'"', '', text)
    # Remove <a href= and "- <a href=" occurrences
    clean_text = re.sub(r'[!"#$%&\'()*+,-./:;<=>?@\[\\\]^_{|}~"]', '', text)
    clean_text = re.sub(r'<a href=|"- <a href="', '', text)
    # Remove square brackets
    clean_text = re.sub(r'[^a-zA-Z0-9\s,.\'"]', '', text)
    clean_text = re.sub(r'\[.*?\]', '', clean_text)
    # Remove other special characters
    clean_text = re.sub(r'[<>\\/[\]]', '', clean_text)
    clean_text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    # Remove emoji placeholders and other unwanted characters
    clean_text = re.sub(r':[^:\s]*(?:::[^:\s]*)*:', '', clean_text)
        # Remove less than and greater than symbols
    clean_text = re.sub(r'[<>]', '', text)
    # Remove extra whitespaces
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    return clean_text

# Apply the remove_emojis function to the "Comment" column
data1['comment'] = data1['comment'].apply(remove_emojis)
```
